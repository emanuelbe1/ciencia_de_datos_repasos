##### <b>N√∫mero de reproducciones (Spotify) y adecuaci√≥n para el baile</b>
<p>Para intentar responder a la primer pregunta, voy a hacer un g≈ïafico de puntos con la variable de bailabilidad('Danceability') y el n√∫mero de stream's. </p>

```python
# Variables que no voy a utilizar en el an√°lisis: Url_spotify, Uri, Url_youtube, y Description
columnas_a_ignorar = ["Url_spotify", "Uri", "Url_youtube", "Description"]
spotifyYoutube_2 = spotifyYoutube.drop(columns=columnas_a_ignorar)

# scatterplot usando Matplotlib
plt.figure(figsize=(14, 8))
plt.scatter(spotifyYoutube_2['Stream'],spotifyYoutube_2['Danceability'], alpha=0.5)
plt.title('Relaci√≥n entre Danceability y Stream en Spotify')
plt.xlabel('N√∫mero de Reproducciones en Spotify')
plt.ylabel('Danceability')
plt.grid(True)

plt.ticklabel_format(axis='x', style='plain', scilimits=(6, 6))

#
plt.show()

danceability_mean = spotifyYoutube_2['Danceability'].mean()
danceability_median = spotifyYoutube_2['Danceability'].median()
danceability_mode = spotifyYoutube_2['Danceability'].mode()[0]
danceability_stream_corr = spotifyYoutube_2[['Danceability', 'Stream']].corr()['Stream'][0]

print(f"Media de la columna 'Danceability': {danceability_mean}")
print(f"Mediana de la columna 'Danceability': {danceability_median}")
print(f"Moda de la columna 'Danceability': {danceability_mode}") 
print(f'Correlaci√≥n entre n√∫mero de "Streams" y "Danceability": {danceability_stream_corr}') 
print("_____" * 15) 
```

#####  <b>Gr√°fico de dispersi√≥n: instrumentalidad por reproducciones en Spotify</b>
 <p>&nbsp;Para responder a la tercer pregunta, he creado una visualizaci√≥n de las reproducciones en spotify y la instrumentalidad de la canci√≥n. Separe las canciones con menos probabilidad de tener contenido vocal (mayores a 0.5 de instrumentalness) en color azul, y con color rojo las que m√°s probabilidad tienen de contener vocalizaciones y no ser instrumentales (menos de 0.5 de instrumentalness). </p>

```python
# Filtrar canciones con alta y baja instrumentalidad
# Esto es: las canciones con m√°s del 70% de probabilidad de ser instrumentales
x_instrumentalness_alta = spotifyYoutube_2[spotifyYoutube_2['Instrumentalness'] > 0.5]['Instrumentalness']
y_streams_instrumentalness_alta = spotifyYoutube_2[spotifyYoutube_2['Instrumentalness'] > 0.5]['Stream']

# Esto es : las canciones con menos de 1% de probabilidad de ser instrumentales
x_instrumentalness_baja = spotifyYoutube_2[spotifyYoutube_2['Instrumentalness'] < 0.5]['Instrumentalness']
y_streams_instrumentalness_baja = spotifyYoutube_2[spotifyYoutube_2['Instrumentalness'] < 0.5]['Stream']

# Crear el gr√°fico combinado
plt.figure(figsize=(10, 6))

# gr√°ficando canciones con alta instrumentalidad
plt.scatter(y=y_streams_instrumentalness_alta, x=x_instrumentalness_alta, label='Alta Instrumentalidad', edgecolor='black', alpha=0.3)

# gr√°ficando canciones con baja instrumentalidad
plt.scatter(y=y_streams_instrumentalness_baja, x=x_instrumentalness_baja, label='Baja Instrumentalidad', edgecolor='red', alpha=0.3)

plt.ticklabel_format(style='plain', axis='y', useOffset=False)
plt.title('Reproducciones de Spotify vs. Instrumentalidad')
plt.xlabel('Instrumentalidad')
plt.ylabel('Reproducciones en Spotify')
plt.legend()
plt.show()


instrumentalness_mean = spotifyYoutube_2['Instrumentalness'].mean()
instrumentalness_median = spotifyYoutube_2['Instrumentalness'].median()
instrumentalness_mode = spotifyYoutube_2['Instrumentalness'].mode()[0]
instrumentalness_stream_corr = spotifyYoutube_2[['Instrumentalness', 'Stream']].corr()['Stream'][0]

print(f"Media de la columna 'Instrumentalness': {instrumentalness_mean}")
print(f"Mediana de la columna 'Instrumentalness': {instrumentalness_median}")
print(f"Moda de la columna 'Instrumentalness': {instrumentalness_mode}") 
print(f'Correlaci√≥n entre n√∫mero de "Streams" e "Instrumentalness": {instrumentalness_stream_corr}') 
print("_____" * 15) 
```

Esta m√©trica indica la probabilidad de que una canci√≥n carezca de contenido vocal.
